```python
class MultiVariateLinearRegression:

  def __init__(self):
    self.weights = None
    self.bias = 0
    self.vector_weights = None
    self.vector_bias = 0
    self.n_iterations = 0 
    
  def fit(self, X, y, learning_rate, n_iters, regularization=None, l1_lambda=0.1, beta=0.9, epsilon=1e-3): # 0.01, 1000
    # Normalize the data
    X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    # Add column of ones as bias term to X feature
    X_norm = np.insert(X_norm, 0, 1, axis=1)
    # Initialize weights and bias
    self.weights = np.random.randn(X_norm.shape[1], y.shape[1])
    self.vector_weights = np.zeros((X_norm.shape[1], y.shape[1]))

    gradient_weights = np.inf
    i = 0
    loss_1 = []
    loss_2 = []
    # Perform gradient descent
    while np.linalg.norm(gradient_weights) > epsilon and i < n_iters:
      # Compute the predicted y
      y_pred = np.dot(X_norm, self.weights) + self.bias
      
      # Compute the difference between actual y and predicted y
      error = y_pred - y

      # calculate loss (for this specific dataset that have two target value Ys)
      current_loss_1 = np.mean((error[:, 0])**2)
      current_loss_2 = np.mean((error[:, 1])**2)
      loss_1.append(current_loss_1)
      loss_2.append(current_loss_2)
      
      # Compute the gradients
      gradient_weights = np.dot(X_norm.T, error) / X_norm.shape[0]
      gradient_bias = np.mean(error)

      i+=1 
      
      # Compute the gradients of weights if l1 regularization is chosen
      if regularization == "l1":
        gradient_weights += l1_lambda * np.sign(self.weights)

      # Compute momentum vectors
      self.vector_weights = beta * self.vector_weights + (1-beta) * gradient_weights
      self.vector_bias = beta * self.vector_bias + (1-beta) * gradient_bias

      # Update the weights and bias with momentum
      self.weights -= learning_rate * self.vector_weights
      self.bias -= learning_rate * self.vector_bias
      if i > n_iters or np.linalg.norm(gradient_weights) <= epsilon:
        break
    self.n_iterations=i
    self.loss = np.column_stack((loss_1, loss_2))
    self.total_loss = [np.mean(loss_1),np.mean(loss_2)]

    return self

  def predict(self, X):
    # Add column of ones as bias term to normalized X feature
    # Normalize the data
    X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    X_norm = np.insert(X_norm, 0, 1, axis=1)

    # Compute the predicted y
    y_pred = np.dot(X_norm, self.weights) + self.bias
    
    return y_pred

```

```python
class MiniBatchMultiVariateLinearRegression:

  def __init__(self):
    self.weights = None
    self.bias = 0
    self.vector_weights = None
    self.vector_bias = 0
    self.n_iterations = 0

    
  def fit(self, X, y, learning_rate, n_iters, batch_size, regularization=None,l1_lambda=0.1, beta=0.9, epsilon=1e-3): # 0.01, 1000, 32

    # Normalize the data
    X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    # Add column of ones as bias term to X feature
    X_norm = np.insert(X_norm, 0, 1, axis=1)
    # Initialize weights and bias
    self.weights = np.random.randn(X_norm.shape[1], y.shape[1])
    self.vector_weights = np.zeros((X_norm.shape[1], y.shape[1]))
    
    gradient_weights = np.inf
    i = 0
    loss_1 = []
    loss_2 = []
    # Perform mini-batch stochastic gradient descent
    # for i in range(n_iters):
    while np.linalg.norm(gradient_weights) > epsilon and i < n_iters:
      i+=1
      for j in range(0, X_norm.shape[0], batch_size):
        X_batch = X_norm[j:j+batch_size]
        y_batch = y[j:j+batch_size]

        # Compute the predicted y
        y_pred = np.dot(X_batch, self.weights) + self.bias
      
        # Compute the difference between actual y and predicted y
        error = y_pred - y_batch
        
        # calculate loss (for this specific dataset that have two target value Ys)
        current_loss_1 = np.mean((error[:, 0])**2)
        current_loss_2 = np.mean((error[:, 1])**2)
        loss_1.append(current_loss_1)
        loss_2.append(current_loss_2)
      
        # Compute the gradients
        gradient_weights = np.dot(X_batch.T, error) / batch_size
        gradient_bias = np.mean(error)

        # Compute the gradients of weights if l1 regularization is chosen
        if regularization == "l1":
          gradient_weights += l1_lambda * np.sign(self.weights)
          
        # Compute momentum vectors
        self.vector_weights = beta * self.vector_weights + (1-beta) * gradient_weights
        self.vector_bias = beta * self.vector_bias + (1-beta) * gradient_bias
        
        # Update the weights and bias with momentum
        self.weights -= learning_rate * self.vector_weights
        self.bias -= learning_rate * self.vector_bias

        if i > n_iters or np.linalg.norm(gradient_weights) <= epsilon:
          break
    self.n_iterations=i
    self.loss = np.column_stack((loss_1, loss_2))
    self.total_loss = [np.mean(loss_1),np.mean(loss_2)]

    return self
        
  def predict(self, X):
    # Add column of ones as bias term to normalized X feature
    # Normalize the data
    X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    X_norm = np.insert(X_norm, 0, 1, axis=1)

    # Compute the predicted y
    y_pred = np.dot(X_norm, self.weights) + self.bias
    
    return y_pred
```

```python
class MiniBatchMultivariateLogisticRegression:
  
  def __init__(self):
    self.weights = None
    self.bias = 0
    self.vector_weights = None
    self.vector_bias = 0
    self.n_iterations = 0

     
  def sigmoid(self, z): # activation function
    # actiovation function sigmoid
    return 1/(1+np.exp(-z))

  def fit(self, X, y, learning_rate, epochs, batch_size, regularization=None, l1_lambda=0.1, beta=0.9, epsilon=1e-3): # 0.01, 100,32
    # Normalize the data
    X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

    X_norm = np.insert(X_norm, 0, 1, axis=1)

    # Initialize weights
    self.weights = np.random.randn(X_norm.shape[1], y.shape[1])

    # Initialize weights vector with zeros
    self.vector_weights = np.zeros((X_norm.shape[1], y.shape[1]))
    
    gradient_weights = np.inf

    # Perform gradient descent
    # for _ in range(epochs):
    i = 0
    while np.linalg.norm(gradient_weights) > epsilon and i < epochs:
      for j in range(0, X_norm.shape[0]-batch_size+1, batch_size):
        X_batch = X_norm[j:j+batch_size]
        y_batch = y[j:j+batch_size]
        
        # Compute the predicted y
        y_pred = self.sigmoid(np.dot(X_batch, self.weights) + self.bias)

        # Compute the difference between actual y and predicted y
        error = y_pred - y_batch
        
        # Compute the gradients
        gradient_weights = np.dot(X_batch.T, error) / batch_size
        gradient_bias = np.mean(error)
        
        i+=1

        # Compute the gradients of weights if l1 regularization is chosen
        if regularization == "l1":
          gradient_weights += l1_lambda * np.sign(self.weights)
          
        # Compute momentum vectors
        self.vector_weights = beta * self.vector_weights + learning_rate * gradient_weights
        self.vector_bias = beta * self.vector_bias + learning_rate * gradient_bias
          
        # Update the weights and bias with momentum
        self.weights -= self.vector_weights
        self.bias -= self.vector_bias

        if i > epochs or np.linalg.norm(gradient_weights) <= epsilon:
          break
    self.n_iterations=i

    return self

  def predict(self, X):
    # Normalize the data
    X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    X_norm = np.insert(X_norm, 0, 1, axis=1)

    # Compute the predicted y
    y_pred = self.sigmoid(np.dot(X_norm, self.weights) + self.bias)

    # Rounds the predictions to the nearest integer for output
    y_pred = np.round(y_pred)
    return y_pred
```

```python
class Metrics:
  #confusion matrix calculation
    def confusion_matrix(y_true, y_pred):
      n_classes = np.max(y_true) + 1
      c_matrix = np.zeros((n_classes, n_classes))
      for c1 in range(n_classes):
          for c2 in range(n_classes):
              #(y==c1)*(yh==c2) is 1 when both conditions are true or 0
              c_matrix[c1, c2] = np.sum((y_true==c1)*(y_pred==c2))
      return c_matrix
 #accuracy
    def accuracy(y_true, y_pred):
        # Calculate the prediction accuracy
        return (y_pred == y_true).sum()/(len(y_true))
 #r2 score calculation
    def r2_score(y_true, y_pred):
        # Calculate the r2 score of the prediction
        avg_y_true = np.mean(y_true)
        RSS = np.dot(y_true - y_pred, y_true - y_pred)
        TSS = np.dot(y_true - avg_y_true, y_true - avg_y_true)
        return 1- RSS/TSS
```
